```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
import matplotlib.pylab as pl
import random
```

```
#dS/dt
def fS(S,I,R,N,t,m,g,b):
  return m*N-b*S*I/N-m*S
```

```
#dI/dt
def fI(S,I,R,N,t,m,g,b):
  return b*S*I/N-m*I-g*I
```

```
#dR/dr
def fR(S,I,R,N,t,m,g,b):
  return g*I-m*R
```

**Euler**


```
def f_euler(S,I,R,N,dt,m,g,b):
    dS = (m*N-b*S*I/N-m*S)*dt
    dI = (b*S*I/N-m*I-g*I)*dt
    dR = (g*I-m*R)*dt

    S = S+dS
    I = I+dI
    R = R+dR
    return S,I,R
```

**Poisson**

```
def f_poisson(S,I,R,N,dt,m,g,b):
  for j in range(1,I+1):
    x = random.random()
    y = random.random()
    if (y<S/N and x<b*dt):
      I += 1
      S -= 1
    x = random.random()
    if (x<g*dt):
      I -=1
      R +=1
    x = random.random()
    if (x<m*dt):
      I -=1
      S +=1
  for k in range(1,R+1):
    x = random.random()
    if (x<m*dt):
      S +=1
      R -=1
  return S,I,R
```

plotar graficos

```
def grafico(S_array,I_array,R_array,N_array,t_array, nome):
  sns.set()
  fig,ax = plt.subplots(figsize=(6,4))
  ax.plot(t_array,S_array, color = 'r', label ='S', lw = 2.0)
  ax.plot(t_array,I_array, color = 'b', label ='I', lw = 2.0)
  ax.plot(t_array,R_array, color = 'y', label ='R', lw = 2.0)
  ax.plot(t_array,N_array, color = 'g', label ='N', lw = 2.0)
  ax.set_title("SIR vital:"+str(nome), fontsize =12, y=1.05)
  ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
  ax.set_xlabel("tempo", fontsize = 10)
  ax.set_ylabel("indivíduos", fontsize = 10)
  fig.tight_layout()
  ax.legend()
  fig.show()
  fig.savefig("SIR_"+str(nome))
```

```
N_passos = 10
dt = 250/N_passos

N_array = []; t_array = []; Se_array = []; Ie_array =[]; Re_array =[]; Sp_array = []; Ip_array =[]; Rp_array =[];
#individuos sucetiveis
S = 900; Se = S; Sp = S
#individuos infectuosos
I = 100; Ie = I; Ip = I
#removidos
R = 0; Re = R; Rp = R
N = S+I+R
#transmitibilidade, texa recuperacao, natalidade/morte
b = 0.1; g = 0.1429; m = 0.004

for it in range(1, N_passos+1):
  Se,Ie,Re =  f_euler(Se,Ie,Re,N,dt,m,g,b)
  Sp,Ip,Rp = f_poisson(Sp,Ip,Rp,N,dt,m,g,b)

  Se_array.append(Se)
  Ie_array.append(Ie)
  Re_array.append(Re)

  Sp_array.append(Sp)
  Ip_array.append(Ip)
  Rp_array.append(Rp)

  N_array.append(N)
  t_array.append(it*dt)

grafico(Se_array,Ie_array,Re_array,N_array,t_array, nome = " Euler")
grafico(Sp_array,Ip_array,Rp_array,N_array,t_array, nome = " Método estocástico")
```
